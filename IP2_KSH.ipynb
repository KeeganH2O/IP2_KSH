```python
def month_change(x):
    m_dict = {
        1: 'Janurary',
        2: 'Februaruy',
        3: 'March',
        4: 'April',
        5: 'May',
        6: 'June',
        7: 'July',
        8: 'August',
        9: 'September',
        10: 'October',
        11: 'November',
        12: 'December'
    }
    return m_dict[x]
```

```python
monthly_avg['month'] = list(map(lambda x: month_change(x), monthly_avg['month']))
#tennis['order']=list(map(lambda x: set_event_order(x), tennis['event']))
```

```python
def toHour(x):
    time = {
        0:'12:00 AM',
        1:'1:00 AM',
        2:'2:00 AM',
        3:'3:00 AM',
        4:'4:00 AM',
        5:'5:00 AM',
        6:'6:00 AM',
        7:'7:00 AM',
        8:'8:00 AM',
        9:'9:00 AM',
        10:'10:00 AM',
        11:'11:00 AM',
        12:'12:00 PM',
        13:'1:00 PM',
        14:'2:00 PM',
        15:'3:00 PM',
        16:'4:00 PM',
        17:'5:00 PM',
        18:'6:00 PM',
        19:'7:00 PM',
        20:'8:00 PM',
        21:'9:00 PM',
        22:'10:00 PM',
        23:'11:00 PM'
    }
    return time[x]
```

```python
hourly_avg['Hour'] = list(map(lambda x: toHour(x), hourly_avg['Hour']))
```
